```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
penguin = sns.load_dataset("penguins")
penguin
```

```python
penguin = penguin.dropna()
penguin
```

```python
penguin["species"].unique()
```

```python
penguin.loc[penguin["island"] == "Torgersen", "island"] = 0
penguin.loc[penguin["island"] == "Dream", "island"] = 1
penguin.loc[penguin["island"] == "Biscoe", "island"] = 2
```

```python
penguin.loc[penguin["species"] == "Adelie", "species"] = 0
penguin.loc[penguin["species"] == "Chinstrap", "species"] = 1
penguin.loc[penguin["species"] == "Gentoo", "species"] = 2
```

```python
penguin["sex"].replace("Male", 0, inplace=True)
penguin["sex"].replace("Female", 1, inplace=True)
penguin
```

```python
penguin["species"] = penguin["species"].astype(int)
penguin["island"] = penguin["island"].astype(int)
penguin.info()
```

```python
y = penguin["species"]
X = penguin.iloc[:, 1:]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
X_train
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree

clf = DecisionTreeClassifier(max_depth=3)
clf.fit(X_train, y_train)
```

```python
plt.figure(figsize=(10, 10))
plot_tree(
    clf,
    filled=True,
    feature_names=penguin.columns[1:],
    class_names=["Adelie", "Chinstrap", "Gentoo"],
)
plt.show()
```
